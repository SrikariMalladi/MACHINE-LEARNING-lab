### 4.Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

with open("data.txt", 'r') as file:
    data = [line.strip().split() for line in file]

x_values = [float(row[0]) for row in data]
y_values = [float(row[1]) for row in data]

fig, ax = plt.subplots()
ax.plot(x_values, y_values)

ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Line Plot from Data File')

plt.show()
```

### 5.Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)

plt.plot(x, y1, color='green', linewidth=2, label='Sine')
plt.plot(x, y2, color='blue', linewidth=3, label='Cosine')
plt.plot(x, y3, color='red', linewidth=1, label='2x Sine')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Multiple Lines with Legends')
plt.legend()
plt.show()
```

### 6.Write a Python program to plot quantities which have an x and y position.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
num_points = 20
x_positions = np.random.rand(num_points) * 10
y_positions = np.random.rand(num_points) * 5   

plt.scatter(x_positions, y_positions, color='orange', marker='o', label='Data Points')  

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Quantities with X and Y Positions')

plt.legend()
plt.grid(True)  
plt.show()
```

### 7.Create plot usingsubplots(),scatter, and some other methods on theAxesobject:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(45)
num_points = 20
x_positions_set1 = np.random.rand(num_points) * 15
y_positions_set1 = np.random.rand(num_points) * 10

x_positions_set2 = np.random.rand(num_points) * 15
y_positions_set2 = np.random.rand(num_points) * 10

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

axes[0].scatter(x_positions_set1, y_positions_set1, color='blue', marker='o', label='Set 1')
axes[0].set_title('Scatter Plot - Set 1')
axes[0].set_xlabel('X Axis Label')
axes[0].set_ylabel('Y Axis Label')
axes[0].legend()

axes[1].scatter(x_positions_set2, y_positions_set2, color='green', marker='s', label='Set 2')
axes[1].set_title('Scatter Plot - Set 2')
axes[1].set_xlabel('X Axis Label')
axes[1].set_ylabel('Y Axis Label')
axes[1].legend()

plt.tight_layout()

plt.show()
```

### 8.Generate random data for x and y and create a scatter plot using Matplotlib.

### 9.Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['Widget A', 'Widget B', 'Widget C', 'Widget D']

np.random.seed(42)
sales_data = np.random.randint(50, 200, len(products))

colors = ['green', 'orange', 'blue', 'red']
plt.bar(products, sales_data, color=colors)

plt.xlabel('Products')
plt.ylabel('Sales (in units)')
plt.title('Sales Comparison of Different Widgets')

for i, value in enumerate(sales_data):
    plt.text(i, value + 5, str(value), ha='center', va='bottom')


plt.show()
```

### 10.Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(45)
data = np.random.normal(0, 1, 1000)  # Mean=0, Standard Deviation=1

plt.hist(data, bins=30, color='lightgreen', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Dataset')

plt.show()
```

### 11.Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

```python
import matplotlib.pyplot as plt

categories = ['Groceries', 'Transportation', 'Dining Out', 'Entertainment', 'Utilities', 'Miscellaneous']

expenses = [200, 100, 50, 30, 40, 20]

explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['orange', 'blue', 'green', 'red', 'purple', 'gray'], explode=explode)

plt.title('Expense Distribution in Daily Life Budget')

plt.show()
```
